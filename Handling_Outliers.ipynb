```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv('bigmart.csv')
```

```python
data.head()
```

# **Univariate**

Here Outliers can be present in Item_weifgt, Item_Visibility, Item_MRP,  Item_Outlet_Sales

```python
plt.rcParams['figure.figsize']= (16,4)

plt.subplot(1,4,1)
sb.boxplot(data['Item_Weight'])
plt.subplot(1,4,2)
sb.boxplot(data['Item_Visibility'])
plt.subplot(1,4,3)
sb.boxplot(data['Item_MRP'])
plt.subplot(1,4,4)
sb.boxplot(data['Item_Outlet_Sales'])
```

# **Bivariate**

```python
plt.subplot(1,3,1)
sb.scatterplot(y=data['Item_Outlet_Sales'], x=data['Item_Weight'])
plt.subplot(1,3,2)
sb.scatterplot(y=data['Item_Outlet_Sales'], x=data['Item_Visibility'])
plt.subplot(1,3,3)
sb.scatterplot(y=data['Item_Outlet_Sales'], x=data['Item_MRP'])
```

# **Handling Outliers**

```python
#For Item_Visibility, Outliers are near max value so left it as it as
```

Capping Outliers for Item_Outlet_Sales, cap values greater than 10000 coz most values are less than 10000

```python
data['Item_Outlet_Sales'].values[data['Item_Outlet_Sales'].values>10000]=10000
```

```python
data[data['Item_Outlet_Sales']>10000]
```

```python
plt.rcParams['figure.figsize']= (16,4)

plt.subplot(1,4,1)
sb.boxplot(data['Item_Weight'])
plt.subplot(1,4,2)
sb.boxplot(data['Item_Visibility'])
plt.subplot(1,4,3)
sb.boxplot(data['Item_MRP'])
plt.subplot(1,4,4)
sb.boxplot(data['Item_Outlet_Sales'])
```

For Bivariate, delete all values >8000. It will remove all outliers in all 3 cases

```python
data=data[~(data['Item_Outlet_Sales']>8000)]
```

```python
plt.subplot(1,3,1)
sb.scatterplot(y=data['Item_Outlet_Sales'], x=data['Item_Weight'])
plt.subplot(1,3,2)
sb.scatterplot(y=data['Item_Outlet_Sales'], x=data['Item_Visibility'])
plt.subplot(1,3,3)
sb.scatterplot(y=data['Item_Outlet_Sales'], x=data['Item_MRP'])
```

Now Data is Outlier Free(Some are there but they wont harm model)
